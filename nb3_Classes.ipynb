## O que são classes?

Uma classe é uma estrutura que serve para **definir novos tipos de dados** (dados estruturados) e é formada por **atributos** (dados/itens) e **métodos** (funcionalidades/operações)

> A imagem abaixo ilustra isso. Um TAD (Tipo Abstrato de Dado) é um conjunto de dados estruturados e as operações que podem ser executadas sobre eles

**As classes em linguagens orientadas a objetos, como Python, são usadas para implementar TADs**

Lembre das listas, strings e dicionários. Cada um desses tipos de dados em Python é implementado em uma classe (`list, str e dict`, respectivamente) e tem um conjunto de operações disponíveis (métodos), exemplo:

* A classe `list` tem os métodos `append, insert, pop, remove, sort,` etc...
* A classe `str` tem os métodos `upper, lower, split`, etc...
* A classe `dict` tem os métodos `keys, values, items`, etc...

## O que são objetos?

> Objetos são variáveis cujo tipo é uma classe

Para entender melhor, **lembre que toda variável é de um tipo**. Exemplo:

```
x = 10 
y = 3.5
n = 'prog2'
b = True 
impares = [11, 21, 31]

print(type(x))
print(type(y))
print(type(n))
print(type(b))
print(type(pares))
print(type(impares))
```

Em linguagens orientadas a objetos, como o tipo de uma variável é definido em uma classe, podemos chamar essas variáveis de objetos

```
print(f'pares: {pares}')
print(f'ímpares: {impares}')
```

Por exemplo, acima temos dois objetos da classe `list` (tipo lista): pares e impares

> **Então, sempre que você ler objeto da classe X, lembre que esse objeto é simplesmente uma variável do tipo X**

Cada variável (objeto) tem seus próprios valores, chamados de **estado do objeto**, exemplo:

* O objeto **pares** tem como valores [456, 20, 30, 22]
* O objeto **impares** tem como valores [11, 21, 31]

> Ao executar o comando `impares.append(55)`, por exemplo, iremos adicionar o valor 55 ao estado do objeto impares, resultando em [11, 21, 31, 55]

## Como criar classes

Podemos criar nossas próprias classes para definir um novo tipo de dado, com itens e operações que podem ser realizadas sobre eles

Vamos ver, como primeiro exemplo, a implementação do tipo de dado abstrato chamado Pilha (*também estudamos esse TAD na disciplina de Algoritmos e Estruturas de dados*).



### Exemplo de implementação de uma Pilha

A pilha é um caso especial de lista que permite acesso apenas ao último item inserido

> Dessa forma, sempre insere itens no topo da pilha e sempre remove do topo do pilha

Definimos no método especial `__init__` os dados que irão compor esse tipo (**atributos**) e depois definimos as operações que podem ser feitas sobre esses dados (**métodos**)


A pilha terá os métodos:
* `__init__` que inicializa a pilha vazia
* **push()** que insere um item no topo da pilha
* **pop()** que retira um item do topo da pilha, retornando o elemento
* **size()** que retorna a quantidade de elementos da pilha
* **isEmpty()** que retorna True, se a pilha estiver vazia
* **clear()** que limpa a pilha, ou seja, volta a condição de pilha vazia
* `__repr__` que mostra os elementos da pilha em forma de string

Regras gerais

---


1. Para definir uma classe em Python usamos o comando `class`
2. Fazemos toda a nossa codificação nos **métodos** que seguem as mesmas regras das funções (`def`)
3. A diferença entre um método e uma função está no fato de que nos métodos o primeiro parâmetro sempre é **self**
4. **self** serve para fazer referência ao objeto que está executando o método ou usando o atributo
5. Existem alguns métodos chamados de **especiais**, os principais são `__init__ e __repr__`, eles tem uma sintaxe diferente, pois tem dois underlines antes e depois no nome (por isso, são chamados de **métodos dunders**) 
6. O `__init__` serve para inicializar os dados do objeto
7. O `__repr__` serve para criar uma string que representa os dados do objeto
8. Os atributos também usam o **self** junto do nome, isso faz com ele seja visível em toda a classe (não é uma variável local aos métodos)

> Observe que sempre que usarmos os atributos e métodos dentro da classe, colocamos `self.` junto de seus nomes, isso porque atributos e métodos são dependentes do objeto que está executando os comandos. Veja nos exemplos abaixo que fora da classe para nos referirmos aos atributos e métodos do objeto, precisamos colocar o nome do objeto junto do nome do atributo e do método. Ou seja, dentro da classe usamos `self.atributo` ou `self.método()`, já fora da classe usamos `objeto.atributo` ou `objeto.método()`




```
class pilha:

  #o método especial __init__ é executado quando criamos um objeto da classe
  #é usado para inicializar os atributos
  def __init__(self):
    self.itens = [] #os dados dessa pilha serão armazenados em uma lista
    print('PILHA CRIADA') #observe que ao criar o objeto, esse print será executado

  #insere item no topo da pilha
  def push(self, item):
    self.itens.append(item)
  
  #remove do topo da pilha
  def pop(self):
    try:
      return self.itens.pop()
    except:
      return None
  
  def size(self): 
    return len(self.itens)
  
  def isEmpty(self):
    if len(self.itens) == 0:
      return True
    else:
      return False
  
  def clear(self):
    self.itens = []

  #o método especial __repr__ é executado quando precisamos representar os dados do objeto em forma de string
  #isso acontece, por exemplo, quando 'printamos' o objeto  
  def __repr__(self):
    return str(self.itens) #transforma a lista em string
```

**Uma vez implementada a classe, podemos executar seu código e usá-lo para criar variáveis do seu tipo (objetos)**

```
#para criar um objeto de uma classe nova, usamos o nome da classe e parênteses
minhapilha = pilha() #executa o init
```

> Observe que ao criar a variável minhapilha, foi 'printada' a frase PILHA CRIADA. Isso aconteceu porque ao criar o objeto, o `__init__` é executado para inicialização dos dados

```
minhapilha.push(58)
minhapilha.push(5)
minhapilha.push(8)
minhapilha.push(18)
```

## Programando com orientação a objetos

> Agora que já sabemos criar classes, vamos usar esse conceito para criar programas orientados a objetos

### Exemplo: Jogo de dados

O sistema consiste de 2 dados e até 11 jogadores. Cada jogador escolhe um valor para apostar entre 2 e 12. 

Após os jogadores informarem suas apostas os dados são lançados. O sistema apresenta o resultado: Se a soma do valor das faces dos dados for igual ao valor de uma das apostas, o sistema informa qual o jogador vencedor, caso nenhum jogador acerte o valor, é informado que o computador venceu.

### Classe Jogador

```
class Jogador: #comando para definir uma nova classe

  #Método inicializador dos objetos da classe Jogador (chamado quando criamos um objeto)
  def __init__(self, n, a):
    self.nome = n 
    try:
      self.aposta = int(a)
    except:
      self.aposta = 0 
      print('Sua aposta não foi registrada. Ela precisa ser um número inteiro (entre 2 e 12)') 
    print(f'Jogador criado >> {self}')

  #métodos que servirão para informar nome e aposta do jogador
  def alterarnome(self):
    self.nome = input('Informe o novo nome do Jogador: ')

  def alteraraposta(self):
    try:
      self.aposta = int(input('Informe a nova aposta do Jogador (entre 2 e 12): '))
    except:
      print(f'A aposta deve ser um número inteiro (entre 2 e 12)')
  
  #Método especial usado quando mandamos imprimir o objeto - print
  def __repr__(self):
    return f'Jogador: {self.nome} Aposta: {self.aposta}'
```

**Testando a classe Jogador**

```
#Criação de objetos da classe Jogador
jogador1 = Jogador('José', 9)
jogador2 = Jogador('Maria', 'a')
```

```
print(jogador1)
print(jogador2)
```

### Classe Dado

```
import random #usaremos esse modulo para chamar o metodo randint

class Dado:

  #cria um novo dado que ainda não foi lançado
  def __init__(self):
    self.face = random.randint(1,6)

  #gira o dado para gerar um numero entre 1 e 6 que será o valor da face que ficou para cima
  def girar(self):
    self.face = random.randint(1,6)
  
  def __repr__(self):
    return f'Face deste dado: {self.face}'
  
```

**Testando a classe Dado**

```
dado1 = Dado()
dado2 = Dado()
```

```
dado1.girar()
print(dado1)
```

```
dado2.girar()
print(dado2)
```

### Classe Jogo

```
class Jogo:

  def __init__(self):
    print('\nJOGO DE DADOS\n')
    self.dado1 = Dado() 
    self.dado2 = Dado() 
    self.jogadores = []
    self.iniciarjogo()
    
  def iniciarjogo(self):
    qtdJogadores = int(input('Quantos jogadores vão participar? '))
    for i in range(qtdJogadores):
      n = input(f'\nInforme o nome do Jogador {i+1}: ')
      a = input(f'Informe a aposta do Jogador {i+1} entre 2 e 12: ')
      self.jogadores.append(Jogador(n, a))
  
  def lancardados(self):
    print('\n\nGirando dados')
    self.dado1.girar()
    print(self.dado1)
    self.dado2.girar()
    print(self.dado2)

  def definirvencedor(self):
    resultado = self.dado1.face + self.dado2.face
    print(f'\n\nResultado: {resultado}')
    for i in self.jogadores:
      if i.aposta == resultado:
        print(i.nome + ' Venceu!')
        break
    else:
      print('Computador venceu!')
```

### Executar o jogo

```
meuJogo = Jogo()
meuJogo.lancardados()
meuJogo.definirvencedor()
```
