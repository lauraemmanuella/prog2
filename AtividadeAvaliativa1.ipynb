# Turma 1

```python
class Produto:

  def __init__(self, n, p):
    self.nome = n
    self.preco = p
  
  def __str__(self):
    return f'{self.nome.upper()} R$ {self.preco}'

  def alterar_preco(self):
    self.preco = float(input(f'Informe o novo preço do produto {self.nome}: '))
```

```python
class Loja:

  def __init__(self):
    self.lista = []
  
  def menu(self):
    while True:
      print('\n1-Cadastrar produto')
      print('2-Visualizar produtos cadastrados')
      print('3-Buscar produto')
      print('4-Atualizar preço de produto')
      print('5-Sair')
      opcao = int(input())
      if opcao == 1:
        nome = input('\nInforme o nome do produto: ')
        preco = float(input('Informe o preço do produto: '))
        produto = Produto(nome, preco)
        self.lista.append(produto)
      elif opcao == 2:
        for i in self.lista:
          print(i)
      elif opcao == 3:
        nome = input('\nInforme o nome do produto: ')
        for i in self.lista:
          if nome.upper() == i.nome.upper():
            print(i)
            break
        else:
          print('\nProduto não encontrado')
      elif opcao == 4:
        nome = input('\nInforme o nome do produto: ')
        for i in self.lista:
          if nome.upper() == i.nome.upper():
            i.alterar_preco()
            break
        else:
          print('\nProduto não encontrado')
      elif opcao == 5:
        break
```

```python
loja = Loja()
loja.menu()
```

# Turma 2

```python
class Conta:

  def __init__(self, n, s):
    self.numero = n
    self.saldo = s
  
  def __str__(self):
    return f'Conta número: {self.numero} Saldo: R$ {self.saldo}'

  def deposito(self, valor):
    self.saldo += valor
  
  def saque(self, valor):
    if self.saldo >= valor:
      self.saldo -= valor
    else:
      print('\nSaldo insuficiente')
```

```python
class Agencia:

  def __init__(self):
    self.lista = []
  
  def menu(self):
    while True:
      print('\n1-Cadastrar conta')
      print('2-Visualizar contas cadastradas')
      print('3-Efetuar depósito')
      print('4-Efetuar saque')
      print('5-Sair')
      opcao = int(input())
      if opcao == 1:
        numero = int(input('\nInforme o número da conta: '))
        saldo = float(input('Informe o saldo inicial da conta: '))
        conta = Conta(numero, saldo)
        self.lista.append(conta)
      elif opcao == 2:
        for i in self.lista:
          print(i)
      elif opcao == 3:
        numero = int(input('\nInforme o número da conta: '))
        for i in self.lista:
          if numero == i.numero:
            valor = float(input('Informe o valor do depósito: '))
            i.deposito(valor)
            break
        else:
          print('\nConta não encontrada')
      elif opcao == 4:
        numero = int(input('\nInforme o número da conta: '))
        for i in self.lista:
          if numero == i.numero:
            valor = float(input('Informe o valor do saque: '))
            i.saque(valor)
            break
        else:
          print('\nConta não encontrada')
      elif opcao == 5:
        break
```

```python
agencia = Agencia()
agencia.menu()
```
