## O que são exceções?


> Exceção é uma **falha na execução de uma instrução** e pode ser evitada por codificação adequada ou pode ser contornada por tratamento de exceções (**que evita que o programa pare sua execução**)

Python implementa várias classes que tratam essas exceções

**Se a exceção não é prevista e tratada pelo programador, o programa termina com uma mensagem de erro**

No exemplo abaixo temos uma divisão entre 2 números lidos do usuário:

````
def divisao():
  num1 = float(input('Informe o primeiro número: '))
  num2 = float(input('Informe o segundo número: '))
  res = num1 / num2

  print(f'{num1} / {num2} = {res}')
````

````
divisao()
````

> No entanto, se ao invés de informar um número, o usuário digitar uma letra, o programa irá finalizar com uma mensagem de erro

````
divisao()
````

> Ou se o usuário digitar 0 para o segundo número, outro erro irá aparecer

````
divisao()
````

## Tratando as exceções

> O programador pode prever algumas dessas possibilidades de exceção e tratar o código para evitar que o programa pare

Cada exceção corresponde a um objeto de alguma classe de exceção pré-definida

> Existem várias classes de exceção, cada uma tratando erros/avisos específicos

Os tipos de Exceções podem ser encontrados na documentação oficial: https://docs.python.org/3/library/exceptions.html


Para tratar possíveis exceções, o que precisamos fazer é colocar o bloco de comandos que pode gerar a exceção dentro de um bloco `try` e colocar o que fazer caso isso ocorra em um bloco `except`

```
try:
 	# código que pode apresentar exceção
except NomeDaClasseDaExcecao:
 	# o que fazer se ocorrer a exceção
```



````
def divisao():
  try:
    num1 = float(input('Informe o primeiro número: '))
    num2 = float(input('Informe o segundo número: '))
    
    res = num1 / num2

    print(f'{num1} / {num2} = {res}')

  except ZeroDivisionError:
    print('Ocorreu um erro, pois não pode dividir por 0')
  except ValueError:
    print('Ocorreu um erro, pois o valor digitado não é um número')
  else:
    print('Tudo Ok')
  finally:
    print('Concluindo processo')

divisao()
````

Os blocos `else e finally` são opcionais. 
* O `else` irá ser executado somente quando não ocorrer exceção
* Já o finally sempre é executado independente se teve ou não exceção


> Vejamos outra forma de tratar a exceção:

````
def divisao():
  try:
    num1 = float(input('Informe o primeiro número: '))
    num2 = float(input('Informe o segundo número: '))
      
    res = num1 / num2

    print(f'{num1} / {num2} = {res}')

  except Exception as e:
    print('Ocorreu um erro, mensagem: ', e)

divisao()
````

> Nesse caso, estamos usando uma exceção genérica (`Exception`) que irá capturar qualquer uma das exceções anteriores

**Isso é usado quando o tratamento a ser dado é o mesmo**


Vamos ver uma terceira forma de fazer o tratamento da exceção no exemplo abaixo.

> Nesse exemplo, colocamos um laço para que o usuário repita a entrada de dados até que esteja tudo ok

Nesse caso, o `else` será executado quando nenhuma exceção for disparada para que o laço termine

````
def divisao():
  while True:
    try:
      num1 = float(input('Informe o primeiro número: '))
      num2 = float(input('Informe o segundo número: '))
      
      res = num1 / num2

      print(f'{num1} / {num2} = {res}')

    except Exception as e:
      print('Ocorreu um erro, mensagem: ', e)

    else:
      break

divisao()
````

## Exercício

Faça um programa que simule uma calculadora para as quatro operações básicas. O programa deve ter um menu com as seguintes opções: (1) soma, (2) subtração, (3) multiplicação, (4) divisão e (5) sair. As operações (1), (2), (3) e (4) devem ser implementadas dentro dos métodos a serem definidos pelo programador. 

O programa termina quando o usuário digita (5) para sair, caso contrário o menu aparece novamente. As exceções devem ser tratadas para os casos onde o usuário digite algo que não possa ser convertido para float, ou tente usar o zero como divisor.

**OBS: Programe a calculadora usando orientação à objetos**

````
class Calculadora:

  def __init__(self):
    self.a = 0
    self.b = 0

  def soma(self):
    print(self.a + self.b)

  def subtracao(self):
    print(self.a - self.b)
  
  def multiplicacao(self):
    print(self.a * self.b)
  
  def divisao(self):
    try:
      c = self.a / self.b
    except ZeroDivisionError:
      print('\nNão pode dividir por zero!')
    else:
      print(c)
    
  def menu(self):
    op = 0
    while op != 5:
      print('\n1-Somar')
      print('2-Substrair')
      print('3-Multiplicar')
      print('4-Dividir')
      print('5-Sair')
      try:
        op = int(input('Informe a operação: '))
        if op == 5:
          break
        self.a = float(input('Informe o primeiro valor: '))
        self.b = float(input('Informe o segundo valor: '))
      except:
        print('\nInforme um número válido')
      else:
        if op == 1:
          self.soma()
        elif op == 2:
          self.subtracao()
        elif op == 3:
          self.multiplicao()
        elif op == 4:
          self.divisao()
````

````
c = Calculadora()
c.menu()
````
