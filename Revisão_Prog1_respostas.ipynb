**Questão 1:** Escreva uma expressão que será utilizada para decidir se um aluno foi ou não aprovado em uma disciplina. Para ser aprovado, a média do aluno deve ser maior ou igual a 7, ou a média deve ser maior ou igual a 5 e o percentual de faltas menor que 50%. 

Receba 3 notas referentes às avaliações da disciplina e o percentual de faltas como entrada. Calcule a média ponderada com pesos 4, 5 e 6 (como mostrado a seguir) e mostre se aprovado é `True ou False`, como no exemplo de execução abaixo:

> `media = (nota1 * 4 + nota2 * 5 + nota3 * 6) / 15`

```python
#Seu código aqui

nota1 = float(input('Informe a nota da primeira avaliação (entre 0 e 10): '))
nota2 = float(input('Informe a nota da segunda avaliação (entre 0 e 10): '))
nota3 = float(input('Informe a nota da terceira avaliação (entre 0 e 10): '))
faltas = float(input('Informe o percentual de faltas na disciplina (entre 0-1): '))

media = (nota1 * 4 + nota2 * 5 + nota3 * 6) / 15
print(f'Média ponderada: {media}')

aprovado = media >=7 or (media >=5 and faltas < 0.5)

print(f'Média ponderada: {media:.2f} \nAprovado: {aprovado}')
```

##2. Estrutura condicional

Revisar a estrutura condicional `(if..elif..else)`

**Questão 2:** O IMC – Índice de Massa Corporal é um critério da Organização Mundial de Saúde para dar uma indicação sobre a condição de peso de uma pessoa adulta.
A fórmula é `IMC = peso (kg) / altura (m) x altura (m)`. Elabore um algoritmo que leia o peso e a altura de um adulto e mostre sua condição de acordo com as informações abaixo:

> Classificação do IMC:

* Menor que 18,5 - Abaixo do peso

* Entre 18,5 e 24,9 - Peso normal

* Entre 25 e 29,9 - Sobrepeso (acima do peso desejado)

* Igual ou acima de 30 - Obesidade





```python
#Seu código aqui

peso = float(input('Informe o peso (kg): '))
altura = float(input('Informe a altura (m): '))

imc = peso / (altura * altura)
print(f'IMC = {imc:.2f}')

if imc < 18.5:
  print('Abaixo do peso normal')
elif imc < 25:
  print('Peso normal')
elif imc < 30:
  print('Acima do peso normal')
else:
  print('Obesidade')
```

##3. Estrutura de repetição

Revisar as estruturas de repetição `(for e while)`

**Questão 3:** Faça um programa que receba um número inteiro positivo e mostre todos os seus divisores, como no exemplo de execução abaixo:

```python
#Seu código aqui

num = int(input('Informe um número natural: '))
print(f'Os divisores de {num} são: ')
div = 1
while div <= num:
  if  num % div == 0:
    print(div)
  div += 1
```

**Questão 4**: Faça um programa que mostre todos os números primos entre 2 e 9

```python
for n in range(2, 10): #para cada numero n entre 2 e 9
  for x in range(2, n): #testa se o numero n é divisivel pelos numeros de 2 até ele mesmo
    if n % x == 0:
      #print(f'{n} = {x} * {n//x}') #se for divisivel, ele não é primo
      break #encerra o laço
  else: #caso ele não tenha entrado do if (saído do laço com o break), ele entrará nesse else 
    print(n, 'é um número primo') #o numero é primo, pois não foi divisível por nenhum numero entre 1 e ele mesmo
```

**Questão 6:** Use for, split() e if para criar uma declaração que imprima as palavras que começam com 's':

```python
st = 'Print only the words that start with S in this sentence'
```

```python
#codigo aqui
for i in st.split():
  if 'S' == i[0].upper(): #upper() colocar a string ou caracter em maiúsculo
    print(i)
```

E se fosse para printar todas as palavras com 's'? 

```python
#codigo aqui
for i in st.split():
  if 'S' in i.upper(): #nesse caso não olha apenas o índice 0, mas toda a string i
    print(i)
```

**Questão 7:** Preencha uma lista com as representações em Fahrenheit da seguintes temperaturas usando compreensão de listas

```python
celsius = [0,10,20.1,34.5]
```

```python
#codigo aqui

#COM COMPREENSÃO DE LISTAS
far = [c * 1.8 + 32 for c in celsius]
print(far)
```

```python
#SEM COMPREENSÃO DE LISTAS
far = []
for c in celsius:
  far.append(c * 1.8 + 32)
print(far)
```

Revisar dicionários

**Questão 9:** Faça um programa que receba o nome de um produto e mostre seu preço. A lista de produtos está no dicionário abaixo:

```python
tabela = {
    'ALFACE': 0.45,
    'BATATA': 1.20,
    'TOMATE': 2.30,
    'CENOURA': 1.50
}
```

```python
#Seu código aqui
while True:
  produto = input('Digite o nome do produto ou fim para encerrrar: ').upper() #o upper colaca a entrada em maiúsculo
  if produto == 'FIM':
    break
  if produto in tabela:
    print(f'Preço {tabela[produto]}')
  else:
    print('Produto não encontrado')
```

Revisar a construção de funções

**Questão 10:** Faça uma função que receba como parâmetro uma frase e retorne uma lista com o tamanho de cada palavra contida na frase:

```python
frase = 'How long are the words in this phrase'
```

```python
#Seu código aqui

def word_lengths(frase):
    lista = frase.split()
    tamanhos = [len(i) for i in lista]
    return tamanhos

print(word_lengths(frase)) #aqui chama a função
```
