# Atividade Robo andando em Ambiente

> Implementação de um Robô andando em um ambiente.

1. Implemente uma classe **Ambiente** que possua como atributos a **quantidade de linhas e quantidade de colunas** do ambiente (tamanho do ambiente)
2. Implemente também uma classe **Robô** que possua como atributos: **nome, ambiente, posiçãoX e posiçãoY**, sendo que a posiçãoX e a posiçãoY indicam a posição do robô no ambiente e devem ser inicializadas com 0. O robô terá as funções: **andar para frente, andar para trás, andar para direita, andar para esquerda**, onde em cada uma dessas funções o robô se locomove em 1 unidade nas linhas ou nas colunas. Faça também um método no robô para **mostrar sua posição**.

* andarFrente >> faz com que o robô ande 1 linha
* andarTras >> faz com que o robô volte 1 linha
* andarDireita >> faz com que o robô ande 1 coluna
* andarEsquerda >> faz com que o robô volte 1 coluna

3. Faça um scrip que inicialize o ambiente e o robo e tenha um controle com as opções mostradas abaixo:

CONTROLE DO ROBO

1. Ver Posiçao do Robô
2. Andar para frente
3. Andar para trás
4. Andar para direita
5. Andar para esquerda
6. Sair


### Classe Ambiente

```
class Ambiente:

  def __init__(self, x, y):
    self.qtdLinhas = x
    self.qtdColunas = y
    print(f'Ambiente criado com {self.qtdLinhas} linhas e {self.qtdColunas} colunas')

  def __str__(self):
    return f'Ambiente com {self.qtdLinhas} linhas e {self.qtdColunas} colunas'
```

### Classe Robo

```
class Robo:

  def __init__(self, n, a):
    self.posX = 0
    self.posY = 0
    self.nome = n
    self.ambiente = a # agregação
    print(f'\nOlá! Meu nome é {self.nome}')
  
  def mostrarPosicao(self):
    print(f'\nEstou na posição ({self.posX} , {self.posY}) do ambiente')
  
  def andarFrente(self):
    if  self.posX + 1 < self.ambiente.qtdLinhas:
      self.posX += 1
      print('\nAndando para frente...')
    else:
      print('\nParede à frente!!')
  
  def andarTras(self):
    if  self.posX - 1 >= 0:
      self.posX -= 1
      print('\nAndando para trás...')
    else:
      print('\nParede atrás!!')

  def virarDireita(self):
    if  self.posY + 1 < self.ambiente.qtdColunas:
      self.posY += 1
      print('\nAndando para direita...')
    else:
      print('\nParede à direita!!')
  
  def virarEsquerda(self):
    if  self.posY - 1 >= 0:
      self.posY -= 1
      print('\nAndando para esquerda...')
    else:
      print('\nParede à esquerda!!')
  
  def __repr__(self):
    return f'\nOlá! Meu nome é {self.nome}, estou na posição ({self.posX} , {self.posY}) do {self.ambiente}'
```

### Executar

```
sala = Ambiente(3, 3)
```

```
mig = Robo('Mig', sala)
```

```
print('CONTROLE DO ROBO')
opcao = 0
while opcao != 6:
  print('\n1-Ver Posiçao do Robô')
  print('2-Andar para frente')
  print('3-Andar para trás')
  print('4-Andar para direita')
  print('5-Andar para esquerda')
  print('6-Sair')

  opcao = int(input('\n>>'))

  if opcao == 1:
    mig.mostrarPosicao()
  elif opcao == 2:
    mig.andarFrente()
  elif opcao == 3:
    mig.andarTras()
  elif opcao == 4:
    mig.virarDireita()
  elif opcao == 5:
    mig.virarEsquerda()
  elif opcao == 6:
    print('\nDesligando controle\n')
  else:
    print('\nNão sei o que fazer!')
```
