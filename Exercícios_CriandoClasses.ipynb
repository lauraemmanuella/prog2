# Exercício 1: Retangulo

Crie uma classe em Python que represente um retângulo

> Um retângulo possui dois atributos: base e altura

Defina métodos para obter a base, obter a altura e calcular a área do retângulo


```
class Retangulo:

  def obterBase(self):
    self.base = float(input('Informe o valor da base do retângulo: '))
  
  def obterAltura(self):
    self.altura = float(input('Informe o valor da altura do retângulo: '))

  def calcularArea(self):
    self.area = self.base * self.altura
    return self.area
```

```
r1 = Retangulo() #Cria objeto do tipo Retangulo
```

```
r1.obterBase() #objeto r1 chama o método obterBase() da classe Retangulo
```

```
r1.obterAltura() #objeto r1 chama o método obterAltura() da classe Retangulo
```

```
r1.calcularArea()#objeto r1 chama o método calcularArea() da classe Retangulo
```

```
from datetime import date

class pessoa:

  def __init__(self, n, c, a, m, d):
    self.nome = n
    self.cpf = c
    self.datanasc = date(a, m, d)

  def cumprimentar(self):
    print(f'Olá, meu nome é: {self.nome}. CPF: {self.cpf}')
  
  def calcular_idade(self):
    idade = (date.today() - self.datanasc)/365
    print(f'Idade: {idade.days}')
  
  def __str__(self):
    return f'Nome: {self.nome}\nCPF: {self.cpf}\nData de nascimento: {self.datanasc}'
  
  def __eq__(self, outraPessoa):
    if self.cpf == outraPessoa.cpf and self.nome == outraPessoa.nome:
      return True
    else:
      return False
```

```
p1 = pessoa('Cristina', '001.002.003-04', 2009, 11, 4)
p2 = pessoa('Cristina', '001.002.003-04', 2009, 11, 4)

p3 = pessoa('João', '003.004.005-06', 2015, 6, 11)
p4 = p1

p1.cumprimentar()
p2.cumprimentar()
p3.cumprimentar()
p4.cumprimentar()
```

# Exercício 3: Ponto

Considere o conceito de um ponto matemático.

Implementar as operações:
*   pedir sua coordenada x (getX)
*   pedir sua coordenada y (getY)
*   calcular a distância de um ponto para a origem

Implemente o método str para mostrar o ponto no formato: (X, Y)


Implemente o método eq para que os objetos sejam considerados iguais quando tiverem o mesmo valor de X e Y

```
import math

class Ponto:

  #Inicia o objeto com os valores passados ou com 0,0 caso nada tenha sido passado
  def __init__(self, xInicial = 0, yInicial = 0):
    self.x = xInicial
    self.y = yInicial

  def getX(self):
    return self.x
  
  def getY(self):
    return self.y

  def distanciaOrigem(self):
    return math.sqrt((self.x ** 2)+(self.y**2))

  def __str__(self):
    return '(' + str(self.x) + ',' + str(self.y)+ ')'

  def __eq__(self, ponto):
    if self.x == ponto.x and self.y == ponto.y:
      return True
    else:
      return False
```

```
p1 = Ponto(2,3)
p2 = Ponto(1,1)
p3 = Ponto(2,3)
```

```
print(p1)
```

```
print(p3)
```

```
print(p1 == p2)
```

```
print(p1 == p3)
```

Essa classe deve possuir métodos que sirvam para acessar e modificar os valores de canal e volume em uma unidade cada vez. 

```
class Televisao:

  qtd = 0

  def __init__(self, c = 1, v = 0):
    self.canal = c
    self.volume = v
    Televisao.qtd += 1

  def maisVolume(self):
    self.volume = self.volume + 1

  def menosVolume(self):
    self.volume = self.volume - 1

  def maisCanal(self):
    self.canal = self.canal + 1

  def menosCanal(self):
    self.canal = self.canal - 1
  
  def __str__(self):
    return 'Tv ligada\ncanal: '+ str(self.canal) + ' volume: ' + str(self.volume)
  
  def __eq__(self, outraTV):
    if self.canal == outraTV.canal:
      return True
    else:
      return False
```
