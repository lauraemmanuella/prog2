```python
import random

class Dado:

  def __init__(self):
    self.face = random.randint(1,6) 

  def girar(self):
    self.face = random.randint(1,6)
    print(f'\nDado: {self.face}')

  def __str__(self):
    return f'\nDado: {self.face}'
```

```python
class Jogador:

  def __init__(self, n):
    self.nome = n
    self.posicao = 0
  
  def __str__(self):
    return self.nome
```

```python
class Tabuleiro:

  def __init__(self, jogadores = None):
    self.casas = [i for i in range(22)]
    self.premios = {0: 'Início',
                    1: 'Avance 3 casas',
                    6: 'Volte 2 casas',
                    8: 'Avance 2 casas',
                    10: 'Volte 3 casas',
                    13: 'Avance 1 casa',
                    19: 'Volte 1 casa',
                    21: 'Chegada' 
                    }
    self.jogadores = jogadores
  
  def premio(self, casa):
    valor = 0
    if casa == 1:
      valor = 3
    elif casa == 6:
      valor = -2
    elif casa == 8:
      valor = 2
    elif casa == 10:
      valor = -3
    elif casa == 13:
      valor = 1
    elif casa == 19:
      valor = -1
    print(self.premios[casa])
    return valor

  def __str__(self):
    out = '\n'
    for i in self.casas:
      premio = ''
      jogadores = ''
      if i in self.premios:
        premio = self.premios[i]
      for j in self.jogadores:
        if j.posicao == i:
          jogadores += ' ' + j.nome    
      out = out + ' ' + str(i) + ' '+ premio + ' '+ jogadores +'\n'
    return out
```

```python
class Jogo:

  def __init__(self, qtdJogadores):
    self.jogadores=[Jogador('J'+str(i)) for i in range(1, qtdJogadores+1)]
    self.tabuleiro = Tabuleiro(self.jogadores)
    self.dado = Dado()

  def jogar(self):
    vencedor = False
    while not vencedor:
      for i in self.jogadores:
        print(self.tabuleiro) 
        input(f'\n{i.nome} girar dado [aperte qualquer tecla]')
        self.dado.girar() 
        novaposicao = i.posicao + self.dado.face
        print(f'\nJogador {i} vai para posição {novaposicao}')
        while novaposicao in self.tabuleiro.premios and novaposicao != 21:
          novaposicao = novaposicao + self.tabuleiro.premio(novaposicao)
          print(f'\nJogador {i} vai para posição {novaposicao}')
        i.posicao = novaposicao
        if i.posicao >= 21:
          print(f'{i} ganhou!!')
          vencedor = True
          break
```

```python
j = Jogo(2)
j.jogar()
```
